```python
from ultralytics import YOLO

# Load YOLOv8 Small model (you can also use 'yolov8n.pt' or 'yolov8m.pt')
model = YOLO('yolov8s.pt')

# Train for 10 epochs
model.train(
    data='/content/drive/MyDrive/NORTHEASTERN/subject/MasterP/Assignments/project/dataset/SODA-A/SODA-A-Processed/data.yaml',
    epochs=10,
    imgsz=640,
    batch=16,
    workers=2,
    project='SODA-A-YOLOv8',
    name='yolov8_run1',
    exist_ok=True
)
```
