```python
import os

# Create the required directories
os.makedirs("/content/drive/MyDrive/NORTHEASTERN/subject/MasterP/Assignments/project/dataset/SODA-A/SODA-A-Processed/labels/train", exist_ok=True)
os.makedirs("/content/drive/MyDrive/NORTHEASTERN/subject/MasterP/Assignments/project/dataset/SODA-A/SODA-A-Processed/labels/val", exist_ok=True)

os.makedirs("/content/drive/MyDrive/NORTHEASTERN/subject/MasterP/Assignments/project/dataset/SODA-A/SODA-A-Processed/images/train", exist_ok=True)
os.makedirs("/content/drive/MyDrive/NORTHEASTERN/subject/MasterP/Assignments/project/dataset/SODA-A/SODA-A-Processed/images/val", exist_ok=True)

print("✅ Folders created if they didn't already exist.")
```

```python
import shutil
import os

# TRAIN
src_train_labels = "/content/drive/MyDrive/NORTHEASTERN/subject/MasterP/Assignments/project/dataset/SODA-A/SODA-A-Processed/labels/train"
dst_train_labels = "/content/drive/MyDrive/NORTHEASTERN/subject/MasterP/Assignments/project/dataset/SODA-A/SODA-A-Processed/images/train/labels"

# VAL
src_val_labels = "/content/drive/MyDrive/NORTHEASTERN/subject/MasterP/Assignments/project/dataset/SODA-A/SODA-A-Processed/labels/val"
dst_val_labels = "/content/drive/MyDrive/NORTHEASTERN/subject/MasterP/Assignments/project/dataset/SODA-A/SODA-A-Processed/images/val/labels"

# Clean up old directories if they exist
for path in [dst_train_labels, dst_val_labels]:
    if os.path.exists(path):
        shutil.rmtree(path)

# Copy label directories
shutil.copytree(src_train_labels, dst_train_labels)
shutil.copytree(src_val_labels, dst_val_labels)

print("✅ Labels copied for YOLO training compatibility.")
```
